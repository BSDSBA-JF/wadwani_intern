# Loading necessary libraries

```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

# MCA Joblist EDA

## Loading the dataset

```python
df = pd.read_csv("MCA JOBLIST.csv")
```

## Shape of Dataset

```python
print("Shape:", df.shape)
```

MCA JOBLIST has 1151 rows x 7 columns

## Checking for any missing values

```python
missing_counts = df.isnull().sum()

missing_percent = (df.isnull().mean() * 100).round(2)

unique_vals = df.apply(lambda x: x.dropna().unique()) # unique values per columnd

missing_df = pd.DataFrame({
    'Missing Values': missing_counts,
    'Percent': missing_percent,
})
missing_df
```

Based on the table above, there are no missing values present in each column of the dataset.

```python
records = []
for col in df.columns:
    vals = sorted(df[col].dropna().unique().tolist())
    records.append({
        "column": col,
        "unique_count": len(vals),
        "unique_values": vals
    })
unique_values_df = pd.DataFrame(records)
unique_values_df 
```

- There are 1151 unique job titles, which matches the number of rows of the dataset
- No duplicates within the job title column
- The dataset uses binary fields, but the labeling conventions differ: some columns use `Yes/No` while others use `Y/N`.


# Univariate Analysis

```python
sns.set(style="whitegrid")

print("\nJob Sector counts:")
print(df['Job Sector'].value_counts())

plt.figure(figsize=(12,6))
sns.countplot(y='Job Sector', data=df, order=df['Job Sector'].value_counts().index)
plt.title("Job Sectors Frequency")
plt.show()
```

**Job Sectors**

A majority of the job sectors concentrated in manufacturing, human health & social work, and professional/scientific/technical Activities. The distribution follows a long-tail pattern, where only a handful of sectors dominate, while many others have relatively few jobs.

```python
print("\nJob Subsector counts:")
print(df['Job Subsector'].value_counts())

plt.figure(figsize=(12,8))
sns.countplot(y='Job Subsector', data=df, order=df['Job Subsector'].value_counts().index[:20])
plt.title("Top 20 Job Subsectors Frequency")
plt.show()
```

**Job subsectors**

Among job subsectors, no single area dominates as heavily as in the sector-level view. The most common subsectors are Creative/Arts/Entertainment, Social Work (without accommodation), Medical and Dental Practice, and R&D in Social Sciences/Humanities. The distribution is flatter compared to sectors, meaning opportunities are spread across a wider range of subsectors rather than being concentrated in a few.

```python
print("\nEducational Pathway distribution:")
print(df['Educational Pathway'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x='Educational Pathway', data=df)
plt.title("Educational Pathway Distribution")
plt.show()
```

**Education Pathway Distribution**

About two-thirds of jobs in the dataset require higher education, while one-third does not. This suggests that non-higher education pathways (such as vocational training or technical skilling) still represent a significant share of opportunities and should not be overlooked in workforce planning.

```python
print("\nPRC requirement (Yes/No):")
print(df['HEI with PRC (Professional Regulation Commission) Exam'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x='HEI with PRC (Professional Regulation Commission) Exam', data=df)
plt.title("PRC Requirement (Yes/No)")
plt.show()
```

**PRC Requirement**

Most jobs in the dataset (86%) do not require PRC licensing, while only 14% do. This means that the majority of opportunities are accessible without regulatory exams, though the smaller PRC-regulated segment remains vital for professions like healthcare and engineering.

```python
print("\nPriority sector (Y/N):")
print(df['Priority Sector (Y/N)'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x='Priority Sector (Y/N)', data=df)
plt.title("Priority Sector (Y/N)")
plt.show()
```

**Priority Sectors**

Around 40% of jobs are flagged as priority. Understanding which sectors fall into this category (likely health, ICT, and manufacturing) is important because it shows where policy and industry training efforts should be focused.

# Bivariate Analysis

```python
ct = pd.crosstab(df["Job Sector"], df["Educational Pathway"], normalize="index") * 100

ct_sorted = ct.sort_values("Higher Education", ascending=False)

ax = ct_sorted.plot(
    kind="barh",
    stacked=True,
    figsize=(12,6),
    color=["#66b3ff", "#ff9999"]  
)

ax.set_ylabel("Percentage of Jobs")
ax.set_xlabel("Job Sector")
ax.set_title("Job Sector  Educational Pathway (row %)")

plt.xlim(0, 100) 

plt.legend(
    title="Educational Pathway",
    bbox_to_anchor=(-0.6, 1.1),  
    loc="upper left"
)

plt.show()
```

**Job sector cross-tabulated with Educational Pathway**

Sectors like Professional, Scientific, and Technical Activities; Public Administration; Education; and Financial/Insurance Activities are almost entirely blue. This means jobs in these sectors overwhelmingly require higher education qualifications (college/university degrees). The sectors that are mixed (Wholesale/Retail, Transport & Storage, etc.) suggest multiple entry routes; not only could they be accessed using degrees, but also technical/vocational training. FIelds marked purely in red do no require higher formal education whatsoever.

```python
ax = ct_sorted.plot(
    kind="barh", stacked=True, figsize=(12,6), color=["#ff9999", "#66b3ff"]
)

ax.set_ylabel("Percentage of Jobs")
ax.set_xlabel("Job Sector")
ax.set_title("Job Sector  Priority Sector (row %)")

plt.xlim(0, 100) 

plt.legend(
    title="Priority Sector",
    labels=["No", "Yes"],
    bbox_to_anchor=(-0.6, 1.1),
    loc="upper left"
)


plt.show()
```

**Job Sector cross-tabulated with Priority Sector**

Sectors that are marked entirely blue (Administrative and Support Service Activities & Financial and Insurance Activities), meaning 100% of the jobs under those sectors are marked as "priority." They stand out as *high-focus areas* for government or industry policy. It is likely that more funding, skills training programs, and job creation initiatives are directed towards those sectors as compared to those that are mixed or fully pink.

```python
ct = pd.crosstab(df["HEI with PRC (Professional Regulation Commission) Exam"],
                 df["Educational Pathway"],
                 normalize="index") * 100

ax = ct.plot(kind="bar", stacked=True, figsize=(8,6), color=["#66b3ff", "#ff9999"])

ax.set_ylabel("Percentage of Jobs")
ax.set_xlabel("PRC Requirement")
ax.set_title("PRC Requirement  Educational Pathway (row %)")

plt.xticks(rotation=0)
plt.legend(
    title="Educational Pathway",
    bbox_to_anchor=(1.02, 1),   
    loc="upper left"
)
plt.show()
```

**PRC requirement cross-tabulated with educational pathway**

Jobs requiring PRC exams universally demand higher education, showing a strict regulationeducation link. By contrast, jobs without PRC requirements are more diverse, spanning both higher education and non-higher education pathways, indicating wider flexibility in entry routes.
